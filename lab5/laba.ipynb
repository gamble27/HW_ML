```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from NN import NNBinaryClassifier
```

```python
df = pd.read_csv("/home/olga/Projects/HW_ML/data/data-nn/data1.csv")
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(x='x1', y='x2', data=df[['x1', 'x2']], hue=df['y'], ax=ax)
```

```python
x = df[['x1','x2']].to_numpy()
y = df['y'].to_numpy()
y = y.reshape(y.shape[0], 1)
```

```python
model = NNBinaryClassifier((2,7,7,3), inner_activation="tanh")
model.fit(x, y, epochs=10000, print_cost=1000, lr=.01)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
model.decision_boundary(x, y, ax)
```

```python
model.save("./model1.pkl")
```
